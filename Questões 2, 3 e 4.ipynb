```python
%matplotlib inline
%config InlineBackend.figure_format='svg'
import numpy as np
import matplotlib.pyplot as plt
```

```python
# arquivos csv para a plotagem dos gráficos
n1, tempo1 = np.loadtxt('CoutingResult.txt', delimiter = '\t', skiprows = 1, unpack = True)
n2, tempo2 = np.loadtxt('QuickResult.txt', delimiter = '\t', skiprows = 1, unpack = True)
n3, tempo3 = np.loadtxt('BucketResult.txt', delimiter = '\t', skiprows = 1, unpack = True)
n4, tempo4 = np.loadtxt('BucketResult2.txt', delimiter = '\t', skiprows = 1, unpack = True)
```

```python
#2º questão a):
import random
import time

def heapify(arr, n, i): 
	largest = i # Initialize largest as root 
	l = 2 * i + 1	 # left = 2*i + 1 
	r = 2 * i + 2	 # right = 2*i + 2 

	if l < n and arr[i] < arr[l]: 
		largest = l 

	if r < n and arr[largest] < arr[r]: 
		largest = r 

	if largest != i: 
		arr[i],arr[largest] = arr[largest],arr[i]
		heapify(arr, n, largest) 

def heapSort(arr): 
	n = len(arr) 

	for i in range(n, -1, -1): 
		heapify(arr, n, i) 

	for i in range(n-1, 0, -1): 
		arr[i], arr[0] = arr[0], arr[i]
		heapify(arr, i, 0) 

def preenche(array, n):
	for i in range(0, n):
		aux = random.randint(1,100)
		#print(aux)
		array.insert(0, aux)

arr = []
preenche(arr, 100)

print ("Vetor original:") 
print (arr)
n = len(arr) 
for i in range(n, -1, -1): 
	heapify(arr, n, i) 
print ("Heap:")
print (arr)
heapSort(arr) 
n = len(arr) 
print ("Vetor ordenado:") 
print (arr)
```

```python
#2º questão c):
v1, exec1 = np.loadtxt('HeapResultCres.txt', delimiter = '\t', skiprows = 1, unpack = True)
v2, exec2 = np.loadtxt('HeapResultDecres.txt', delimiter = '\t', skiprows = 1, unpack = True)
v3, exec3 = np.loadtxt('HeapResultAle.txt', delimiter = '\t', skiprows = 1, unpack = True)

plt.plot(v1, exec1, label="ordem crescente")
plt.plot(v2, exec2, label="ordem decrescente")
plt.plot(v3, exec3, label="numeros aleatorios")
plt.title("Comparacao do Heapsort")
plt.xlabel("tamanho da sequencia")
plt.ylabel("tempo")
plt.legend()
plt.grid()
plt.savefig('questao2c.png')

print("""O tempo de execução do algoritmo mantém-se praticamente o mesmo,
apesar do modo como os números estavam inicialmente na sequẽncia.""")
```
